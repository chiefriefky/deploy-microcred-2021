### VIsualisasi Data Sebagai Eksplorasi

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore') 
f,ax=plt.subplots(1,2,figsize=(18,8))
opini['Opini'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_xlabel('Perhitungan Opini Masyarakat')
ax[0].set_ylabel('Count')
sns.countplot('Opini',data=opini,ax=ax[1])
ax[1].set_xlabel('Perhitungan Opini Masyarakat')
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Opini',y='pcr',data=opini)
plt.subplot(2,2,2)
sns.violinplot(x='Opini',y='bisnis',data=opini)
```

```python
opini['Opini'] = opini['Opini'].astype('category')
#df.head()
list1=list()
mylabels=list()
for gen in opini.Opini.cat.categories:
    list1.append(opini[opini.Opini==gen].pcr)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.legend(loc="upper right")
plt.title('Stacked-histogram ')
```

```python
opini['Opini'] = opini['Opini'].astype('category')
#df.head()
list1=list()
mylabels=list()
for gen in opini.Opini.cat.categories:
    list1.append(opini[opini.Opini==gen].bisnis)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.legend(loc="upper right")
plt.title('Stacked-histogram ')
```

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxplot(x='Opini',y='pcr',data=opini,order=['Negative','Neutral','Positive'],linewidth=2.5,orient='v',dodge=False)
```

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxplot(x='Opini',y='bisnis',data=opini,order=['Negative','Neutral','Positive'],linewidth=2.5,orient='v',dodge=False)
```
